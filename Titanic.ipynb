```python
iloc4 = train['Survived']==1
iloc5 = train['Survived']==0
sur_class1 = train[iloc4]
sur_class2 = train[iloc5]
```

# classwise visualization

```python
x = sur_class1['Pclass'].reshape(-1,1)
y = sur_class2['Pclass'].reshape(-1,1)
plt.style.use('ggplot')
plt.figure(figsize = (5,5))
plt.hist([x,y],color = ['r','b'],alpha = 0.7,label = ['survived','not survived'])
plt.title('Classwise passenger survived stats \n 1->upper , 2->middle , 3->lower')
plt.xticks([1,2,3],['upper','middle','lower'])
plt.legend(loc = 'upper left')
```

# Farewise visualization

```python
x1 = sur_class1['Fare'].reshape(-1,1)
y1 = sur_class2['Fare'].reshape(-1,1)
#plt.style.use('fivethirtyeight')
plt.figure(figsize = (5,5))
plt.hist([x1,y1],color = ['r','b'],alpha = 1,label = ['survived','not survived'],histtype ='step')
plt.title('Classwise passenger survived stats ')
plt.legend(loc = 'upper right')
```

```python
label = train['Survived']
train = train.iloc[:,2:]
train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,label,random_state = 0)
clf = DecisionTreeClassifier().fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
imp_ind = pd.Series(clf.feature_importances_)
imp_ind
```
